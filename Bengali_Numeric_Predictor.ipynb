```python
indices=list(range(len(X_train_all)))
np.random.seed(42)
np.random.shuffle(indices)

## Splitting the set of training images to training and validation (80/20 split)
ind=int(len(indices)*0.80)

# training dataset
X_train=X_train_all[indices[:ind]] 
Y_train=y_train_all[indices[:ind]]

# validation dataset
X_val=X_train_all[indices[-(len(indices)-ind):]] 
Y_val=y_train_all[indices[-(len(indices)-ind):]]
```

```python
## Building the 3 layer Conv Model
def BenModel(input_shape):
    
    
    # Define the input placeholder as a tensor with shape input_shape. Think of this as your input image!
    X_input = Input(input_shape)

    # Zero-Padding: pads the border of X_input with zeroes
    X = ZeroPadding2D((3, 3))(X_input)

    ###############################Conv Layer 1######################################
    
    # CONV -> BN -> RELU Block applied to X
    X = Conv2D(8, (5, 5), strides=(1, 1), name='conv0')(X)
    X = BatchNormalization(axis=3, name='bn0')(X)
    X = Activation('relu')(X)

    # MAXPOOL
    X = MaxPooling2D((2, 2), name='max_pool0')(X)
    
    #####################################################################
    
    ###############################Conv Layer 2######################################
    
    # CONV -> BN -> RELU Block applied to X
    X = Conv2D(16, (3, 3), strides=(1, 1), name='conv1')(X)
    X = BatchNormalization(axis=3, name='bn1')(X)
    X = Activation('relu')(X)

    # MAXPOOL
    X = MaxPooling2D((2, 2), name='max_pool1')(X)
    
    #####################################################################
    
    ##############################Conv Layer 3#######################################

    # CONV -> BN -> RELU Block applied to X
    X = Conv2D(32, (3, 3), strides=(1, 1), name='conv2')(X)
    X = BatchNormalization(axis=3, name='bn2')(X)
    X = Activation('relu')(X)

    # MAXPOOL
    X = MaxPooling2D((2, 2), name='max_pool2')(X)

    #################################Flattening Layer####################################
    
    # FLATTEN X (means convert it to a vector) + FULLYCONNECTED
    X = Flatten()(X)
    
    ###############################Final 2 Dense Layers######################################
    X = Dense(64)(X)
    X = Dense(10, activation='softmax', name='fc')(X)
    #####################################################################
    
    # Create model. This creates your Keras model instance, you'll use this instance to train/test the model.
    model = Model(inputs=X_input, outputs=X, name='BenModel')

    return model
```

```python
### Assigning Conv Model
BModel = BenModel(X_train.shape[1:])
```

```python
### Compiling the Conv Model
BModel.compile('adam', 'categorical_crossentropy', metrics=['accuracy'])
```
